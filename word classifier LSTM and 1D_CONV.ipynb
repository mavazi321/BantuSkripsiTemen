```python
import librosa
import keras
import os
import numpy as np
```

```python
def extract_feature(y, sr, frame_persecond=40):
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=20, hop_length=int(sr/10))
    return mfcc.T
    
```

```python
data = []
labels = []
```

```python
def get_data(folder="data"):
    dirs = os.listdir(folder)
    counter = 0
    for dr in dirs:
        for file in os.listdir(folder+"/"+dr):
            if file[-4:]==".wav":
                y, sr = librosa.load(folder+"/"+dr+"/"+file)
                data.append(extract_feature(y, sr))
                labels.append(dr)
                counter += 1
                print("succes load "+str(counter)+" files")
```

```python
get_data()
```

```python
labels_encoded = []
for i in labels:
    if i == "BUKA":
        labels_encoded.append(1)
    else :
        labels_encoded.append(0)
```

```python
data_np = keras.preprocessing.sequence.pad_sequences(data, maxlen=None, dtype='int32', padding='pre', truncating='pre', value=0.0)
labels_encoded = np.vstack(labels_encoded)
```

```python
data_np.shape
```

```python
length = 1208
num_train = 700
index = np.random.randint(0, length, size=length)
train_X = data_np[index[0:num_train]]
train_Y = labels_encoded[index[0:num_train]]
test_x = data_np[index[num_train:]]
test_y = labels_encoded[index[num_train:]]
```

```python
train_X.shape
```

```python
def create_model_lstm():
    model = keras.models.Sequential([
            keras.layers.LSTM(20, return_sequences=True, input_shape=[None, 20]),
            keras.layers.LSTM(20),
            keras.layers.Dense(1, activation='sigmoid')
            ])
    return model
```

```python
def create_model_1DCONV():
    model = keras.models.Sequential([
            keras.layers.Conv1D(filters=20, kernel_size=4, strides=1, padding="valid", input_shape=[None, 20]),
            keras.layers.Conv1D(filters=20, kernel_size=4, strides=1, padding="valid"),
            keras.layers.GlobalMaxPooling1D(),
            keras.layers.Dense(1, activation='sigmoid')
            ])
    return model
```

```python
model = create_model_1DCONV()
optimizer = keras.optimizers.Adam(lr=0.01)
model.compile(loss="binary_crossentropy", optimizer= optimizer, metrics=["acc"])
```

```python
history = model.fit(train_X, train_Y, epochs=100, validation_data = (test_x, test_y))
```
