```python
import librosa
import keras
import os
import numpy as np
import scipy.io.wavfile as wav
import python_speech_features
from keras.utils import to_categorical
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn import metrics
```

```python
data = []
labels = []
```

```python
def extract_data_test(folder="data_yafi"):
    dirs = os.listdir(folder)
    counter = 0
    for dr in dirs:
        for file in os.listdir(folder+"/"+dr):
            if file[-4:]==".wav":
                #load wav
                sfreq, audio = wav.read(folder+"/"+dr+"/"+file)
                
    
                #normalisasi amplitudo
                #audio_x = 2.*(audio - np.min(audio))/np.ptp(audio)-1
                #mfcc
                koef_mfcc = 7
                audio_n = python_speech_features.base.mfcc(audio,samplerate=16000,winlen=0.02,winstep=0.01,numcep=koef_mfcc,nfft=512,preemph=0.95,winfunc=np.hamming)
                data.append(audio_n)
                labels.append(dr)
                print("loaded "+str(counter))
                counter+=1
```

```python
extract_data_test()
```

```python
labels_encoded = []
for i in labels:
    if i == "nyala":
        labels_encoded.append(0)
    elif i=="mati":
        labels_encoded.append(1)
    else :
        labels_encoded.append(2)
data_np = keras.preprocessing.sequence.pad_sequences(data, maxlen=None, dtype='int32', padding='pre', truncating='pre', value=0.0)
labels_encoded = np.vstack(labels_encoded)
labels_encoded = to_categorical(labels_encoded)
```

```python
length = 9460
num_train = 6000
index = np.random.randint(0, length, size=length)
train_X = data_np[index[0:num_train]]
train_Y= labels_encoded[index[0:num_train]]
test_X= data_np[index[num_train:]]
test_Y= labels_encoded[index[num_train:]]
```

```python
def create_model_lstm():
    model = keras.models.Sequential([
            keras.layers.LSTM(15, return_sequences=True, input_shape=[None, 7]),
            keras.layers.LSTM(15),
            keras.layers.BatchNormalization(),
            keras.layers.Dense(3, activation='softmax')
            ])
    return model
```

```python
model = create_model_lstm()
optimizer = keras.optimizers.Adam(lr=0.01)
model.compile(loss="categorical_crossentropy", optimizer= optimizer, metrics=["acc"])
model.summary()
```

```python
history = model.fit(train_X, train_Y, epochs=25, validation_data = (test_X, test_Y))
```

```python
history_dict = history.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
epochs = range(1, len(loss_values) + 1)
plt.plot(epochs, loss_values, 'bo', label='Training loss')
plt.plot(epochs, val_loss_values, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.clf()
acc_values = history_dict['acc']
val_acc_values = history_dict['val_acc']
plt.plot(epochs, acc_values, 'bo', label='Training acc')
plt.plot(epochs, val_acc_values, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
def get_label(plain_label):
    labels = ["nyala", "mati", "lain"]
    result = []
    for label in plain_label:
        result.append(labels[np.argmax(label)])
    return result
```

```python
true_train_list = []
pred_train_list = []
true_validasi_list = []
pred_validasi_list = []
```

```python
length_cross = 9458

index_cross = np.random.randint(0, length_cross, size=length_cross)
cross_data = data_np[index_cross]
cross_label = labels_encoded[index_cross]

k=10
num_validation_samples = len(cross_data) // k
validation_scores = []
for fold in range(k):
    print("FOLD  :"+str(fold))
    validation_data = cross_data[num_validation_samples * fold:num_validation_samples * (fold + 1)]
    validation_data_y = cross_label[num_validation_samples * fold:num_validation_samples * (fold + 1)]
    training_data = np.vstack((cross_data[:num_validation_samples * fold], cross_data[num_validation_samples * (fold + 1):]))
    training_data_y = np.vstack((cross_label[:num_validation_samples * fold], cross_label[num_validation_samples * (fold + 1):]))
    cross_model = create_model_lstm()
    optimizer = keras.optimizers.Adam(lr=0.01)
    cross_model.compile(loss="categorical_crossentropy", optimizer= optimizer, metrics=["acc"])
    cross_model.fit(x = training_data, y = training_data_y, epochs=5)
    
    train_label = cross_model.predict(training_data)
    validasi_label = cross_model.predict(validation_data)
    train_label = get_label(train_label)
    validasi_label = get_label(validasi_label)
    pred_train_list.append(train_label)
    pred_validasi_list.append(validasi_label)
    
    true_train_list.append(get_label(training_data_y))
    true_validasi_list.append(get_label(validation_data_y))
    
    validation_score = cross_model.evaluate(x=validation_data, y = validation_data_y)
    validation_scores.append(validation_score[1])
```

```python
for i in range(0, 10):
    print("\n")
    print("FOLD "+str(i+1)+":")
    print("--TRAINING--")
    print("nyala, mati, lain")
    print(confusion_matrix(true_train_list[i], pred_train_list[i] ))
    print(metrics.classification_report(true_train_list[i], pred_train_list[i], digits=3))
    
    print("--VALIDATION--")
    print("nyala, mati, lain")
    print(confusion_matrix(true_validasi_list[i], pred_validasi_list[i] ))
    print(metrics.classification_report(true_validasi_list[i], pred_validasi_list[i], digits=3))
```
